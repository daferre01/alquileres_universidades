```python
def parsear_inmueble(id_inmueble):

    print( '\n Casa numero: ' + str(id_inmueble).replace("[", "").replace("]", ""))
    url = "https://www.idealista.com/inmueble/" + str(id_inmueble).replace("[", "").replace("]", "") + "/"
    print(url)
    browser.get(url)
    try:
        browser.find_element("xpath",'//*[@id="didomi-notice-agree-button"]').click()
    except:
        pass
    html = browser.page_source
    soup = bs(html, 'lxml')
    try:
        titulo = soup.find('span',{'class':'main-info__title-main'}).text
        print('Titulo: ' + titulo)
        casas['titulo'] = titulo
    except:
        casas['titulo'] = 0
        pass
    try:
        localizacion = soup.find('span',{'class':'main-info__title-minor'}).text.split(',')[0]
        print('Localizacion:'+localizacion)
        casas['localizacion'] = localizacion
    except:
        localizacion=0
        casas['localizacion'] = localizacion
        pass
    try:
        ciudad = soup.find('span',{'class':'main-info__title-minor'}).text.split(',')[1]
        print('Ciudad:'+ciudad)
        casas['ciudad']=ciudad
    except:
        casas['ciudad']=localizacion
        pass
    try:
        precio = int(soup.find('span',{'class':'txt-bold'}).text.replace('.',''))
        print(f'Precio:{precio}')
        casas['precio'] = precio
    except:
        casas['precio'] = 0
        pass
    from bs4 import BeautifulSoup

# Suponiendo que 'soup' es tu objeto BeautifulSoup
    info_div = soup.find('div', {'class': 'info-features'})

    # Verificar si se encontró la etiqueta div con la clase 'info-features'
    if info_div:
        # Encontrar cada span individualmente y obtener su texto
        try:
            metros_cuadrados = info_div.find('span').text.strip()
            casas['m2']=metros_cuadrados
            print("Metros cuadrados:", metros_cuadrados)
        except: 
            pass
        try:
            habitaciones = info_div.find_all('span')[1].text.strip()
            casas['habitaciones']=habitaciones
            print("Habitaciones:", habitaciones)
        except:
            casas['habitaciones']=0
            pass
    else:
        print("No se encontró la etiqueta div con la clase 'info-features'")

    # casas['titulo'] = titulo
    

        
    # casas['caracteristicas_basicas'] = caract_basicas
    df_casas = pd.DataFrame(casas)
    
    return(df_casas.T)
```
