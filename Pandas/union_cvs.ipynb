```python
import pandas as pd 
usc=pd.read_excel('universidades/usc.xlsx')
uvigo=pd.read_excel('universidades/uvigo.xlsx')
udc= pd.read_excel('universidades/udc.xlsx')
```

```python
df_usc = usc[['CAMPUS', 'Facultade','2021/2022']]
df_usc['Universidad']='USC'
df_usc = df_usc.rename(columns={'2021/2022': 'CANTIDAD','Facultade': 'FACULTADE','Universidad':'UNIVERSIDAD'})
df_usc
```

```python
df_udc = udc[['Campus', 'Estudo','Total']]
df_udc['Universidad']='UDC'
df_udc = df_udc.rename(columns={'Total': 'CANTIDAD','Campus': 'CAMPUS','Universidad':'UNIVERSIDAD','Estudo':'FACULTADE'})
df_udc
```

```python
df_uvigo = uvigo[['Campus', 'Centro','Total']]
df_uvigo['Universidad']='UVIGO'
df_uvigo = df_uvigo.rename(columns={'Total': 'CANTIDAD','Campus': 'CAMPUS','Universidad':'UNIVERSIDAD','Centro':'FACULTADE'})
df_uvigo
```

```python
df_galicia=pd.concat([df_uvigo, df_usc, df_udc], axis=0)
```

```python
df_galicia['CAMPUS']=df_galicia['CAMPUS'].str.upper()
df_galicia['UNIVERSIDAD']=df_galicia['UNIVERSIDAD'].str.upper()
```

```python
df_galicia
```

```python
# Eliminar filas en blanco
df_galicia = df_galicia.dropna(axis='rows')
df_galicia
```

```python
df_galicia('universidades/galicia_universidades.xlsx')
```
